```python
#ab_star() function
def ab_star(a:str ,b:str) -> tuple:
    temp=''
    for x,y in zip(a,b):
        if x==y:
            temp+=x
        else:
            break
    return a.strip(temp),b.strip(temp)

#simplified_lev() function
def simplified_lev(a:str ,b:str) -> int:
    temp=''
    for x,y in zip(a,b):
        if x==y:
            temp+=x
        else:
            break
    a_uncom=a.strip(temp)
    b_uncom=b.strip(temp)
    return max(len(a_uncom),len(b_uncom))

import os

print('Fancy Spell-Checker')
inputs=input('File to check: ')     #type 'essay.txt' in the output
while os.path.exists(inputs)==False:
    print('File not found! Please try again...')
    inputs=input('File to check: ')
    
spelling=input('File with known words [enter for default]: ')
if spelling=='':
    spelling='spell.txt'
while os.path.exists(spelling)==False:
    print('File not found! Please try again...')
    spelling=input('File with known words [enter for default]: ')
    if spelling=='':
        spelling='spell.txt'

mistake=input('File with common mistake [enter for default]: ')
if mistake=='':
    mistake='common.txt'
while os.path.exists(mistake)==False:
    print('File not found! Please try again...')
    mistake=input('File with known words [enter for default]: ')
    if mistake=='':
        mistake='common.txt'

def main():
    print('Processing ',inputs)
    
    f1=open(spelling, 'r')
    k_spell=list(l.strip() for l in f1)     #reading the file and storing data in list
    
    f2=open(inputs, 'r')
    lines=f2.readlines()
    
    
    new_lines=[]
    for i in range(0,len(lines)):           #accessing data line by line for further computation
        line=lines[i].strip()
        words=line.split()
    
        mis_word=[]     #list to store spelling mistake
        rep_word=[]     #list to store repetative words
        
        #update_words=''
        for w in words:
            if w.lower() not in k_spell:     #spelling mistake
                mis_word.append(w)
            if words.count(w)>1:             #repetative words
                rep_word.append(w)
            
        if len(mis_word)>0 or len(rep_word)>0:     #condition, when mistakes found...
            wd=[]
            #print line no. and the mistakes
            for j in range(0,len(words)):
                if words[j] in mis_word:
                    if j==0:
                        print(f"Line {i+1}: '{words[j]}' {words[j+1]}")
                    elif j==(len(words)-1):
                        print(f"Line {i+1}: {words[j-1]} '{words[j]}'")
                    else:
                        print(f"Line {i+1}: {words[j-1]} '{words[j]}' {words[j+1]}")
                        
                    
            #print line no. and the repeated words
            for k in range(len(words)):
                if words[k] not in rep_word:
                    wd.append(words[k])
                elif words[k] in rep_word and words[k]==words[k+1]:
                    if k==0:
                        print(f"Line {i+1}: '{words[k]} {words[k+1]}' {words[k+2]}")
                        print(f"(0){words[k]} {words[k+1]} (1)remove")
                    elif k==(len(words)-1):
                        print(f"Line {i+1}: {words[k-2]} '{words[k-1]} {words[k]}'")
                        print(f"(0){words[k]} {words[k+1]} (1)remove")
                    else:
                        print(f"Line {i+1}: {words[k-1]} '{words[k]} {words[k+1]}' {words[k+2]}")
                        print(f"(0){words[k]} {words[k+1]} (1)remove")
                
                    
                    while True:
                        o=input('Option: ')
                        if o=='0':
                            wd+=[words[k],words[k+1]]
                            break
                        elif o=='1':
                            wd.append(words[k])
                            break
                        else:
                            print('Invalid input! Please try again')
            update_words=wd
    
        else:
            update_words=words
        new_lines.append(update_words)

    #writing new texts in file after correcting wrong spellings and repetitive words
    new_file=input('File to save updated document: ') #write 'new_essay.txt' in output
    f3=open(new_file, 'w')
    for i in new_lines:
        t=' '.join(i)
        f3.write(t+'\n') 
    
    f3.close()
    f2.close()
    f1.close()
    recheck=input("Hit enter to continue with another file or type 'quit' exit: ")
    if recheck.lower()!='quit':
        main()


main()
```
